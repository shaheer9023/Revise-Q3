```python
from google.colab import userdata
import google.generativeai as genai

# Get the API key from userdata
API_KEY = userdata.get("GOOGLE_API_KEY")

# Configure the genai client
genai.configure(api_key=API_KEY)
```

```python
model=genai.GenerativeModel("gemini-1.5-flash")
```

```python
import requests
from PIL import Image
from io import BytesIO
from IPython.display import display

# Replace with the URL of your image
url = 'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQ1y725EiCoIpNLBrb_73ryegGqJp31N4t0Oklve286xYphluCEZfmwCZwR0NZksKY67mw&usqp=CAU'

# Fetch the image from the URL
response = requests.get(url)

# Open the image using PIL
img = Image.open(BytesIO(response.content))

# Display the image inline in Jupyter Notebook
display(img)
```

```python
prompt=input("enter prompt : ")
response = model.generate_content([prompt, img])
print(response.text)
```
