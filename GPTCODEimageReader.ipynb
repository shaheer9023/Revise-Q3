```python
from google.colab import userdata
key=userdata.get('2.O_exp')
```

```python
# Gemini API ko install karein
!pip install google-generativeai -q

# Required libraries ko import karein
import google.generativeai as genai
from PIL import Image

# Gemini API key daalein (Yahan apna API key daalein)
API_KEY = key

# Gemini API ko configure karein
genai.configure(api_key=API_KEY)

# Gemini model ko initialize karein
model = genai.GenerativeModel('gemini-2.0-flash-exp')

# Image ka path daalein (Yahan apni image ka path daalein)

from google.colab import files

# File upload karne ke liye
uploaded = files.upload()

# Uploaded file ka naam aur content display karein
for filename in uploaded.keys():
    print(f'File "{filename}" uploaded successfully!')

from IPython.display import Markdown
image_path = filename  # Example: '/content/example.jpg'

# Image ko load karein
img = Image.open(image_path)
prompt=input("enter prompt : ")
# Gemini model ko image aur prompt provide karein
response = model.generate_content([prompt, img])

# Extracted text ko print karein
Markdown( response.text)
```
